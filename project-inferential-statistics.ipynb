```python
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set_style("whitegrid")
sns.FacetGrid(df,hue='survival',height=6).map(sns.distplot,"no_of_nodes").add_legend()

plt.show()
```

```python
import numpy as np
count,bins_count=np.histogram(df,bins=10)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(bins_count[1:],cdf,label="CDF")
plt.legend()
plt.show()
```
